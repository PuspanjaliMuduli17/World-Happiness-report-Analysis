# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

**setting the general style  of the plots**

```python
sns.set_style('darkgrid')
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(10,7)
plt.rcParams['figure.facecolor']='#FFE5B4'
```

```python
data=pd.read_csv('2021.csv')
data
```

```python
data.head()
```

```python
data_columns=['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Perceptions of corruption','Generosity']
```

```python
data=data[data_columns].copy()
```

```python
data
```

```python
df=data.rename(columns={'Country name':'Country_name','Healthy life expectancy':'Healthy_life_expectancy','Regional indicator':'Regional_indicator','Ladder score':'Happiness_score','Logged GDP per capita':'Logged_GDP_per_capita','Social support':'Social_support','Freedom to make life choices':'Freedom_to_make_life_choices','Perceptions of corruption':'Perceptions_of_corruption'})
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Plot between happiness and GDP
plt.rcParams['figure.figsize']=(15,7)
plt.title('Plot between Happiness score and GDP')
sns.scatterplot(x=df.Happiness_score,y=df.Logged_GDP_per_capita,hue=df.Regional_indicator,s=200)
plt.legend(loc='upper left',fontsize='10')
plt.xlabel("Happiness_score")
plt.ylabel("GDP_per_capita")
```

```python
gdp_region=df.groupby('Regional_indicator')['Logged_GDP_per_capita'].sum()
gdp_region
```

```python
gdp_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP by Region')
plt.ylabel('')
```

```python
total_country=df.groupby('Regional_indicator')[['Country_name']].count()
print(total_country)
```

```python
#correlation Map
numerical_df=df.select_dtypes(include=['number'])
cor=numerical_df.corr(method='pearson')
f,ax=plt.subplots(figsize=(10,5))
sns.heatmap(cor,mask=np.zeros_like(cor,dtype=np.bool),cmap="Blues",square=True,ax=ax)
```

```python
#corruption in regions
corruption=df.groupby('Regional_indicator')[['Perceptions_of_corruption']].mean()
corruption
```

```python
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of corruption in various Regions')
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation = 30,ha='right')
plt.bar(corruption.index,corruption.Perceptions_of_corruption)
```

```python
top_10=df.head(10)
bottom_10=df.tail(10)
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))
plt.tight_layout(pad=2)
xlabels=top_10.Country_name
axes[0].set_title('Top 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels,rotation=45,ha='right')
sns.barplot(x=top_10.Country_name,y=top_10.Healthy_life_expectancy,color='orange',ax=axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life expectancy')

xlabels=bottom_10.Country_name
axes[1].set_title('Bottom 10 happiest countries Life Expectancy')
axes[1].set_xticklabels(xlabels,rotation=45,ha='right')
sns.barplot(x=bottom_10.Country_name,y=bottom_10.Healthy_life_expectancy,ax=axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life expectancy')
```

```python
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x=df.Freedom_to_make_life_choices,y=df.Happiness_score,hue=df.Regional_indicator,s=200)
plt.legend(loc='upper left',fontsize='12')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness score')
```

```python
country=df.sort_values(by='Perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Countries with most perception of corruption')
plt.xlabel('Country',fontsize=13)
plt.ylabel("corruption index",fontsize=13)
plt.xticks(rotation=30,ha='right')
plt.bar(country.Country_name,country.Perceptions_of_corruption)
```

```python
country=df.sort_values(by='Perceptions_of_corruption').tail(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Countries with most perception of corruption')
plt.xlabel('Country',fontsize=13)
plt.ylabel("corruption index",fontsize=13)
plt.xticks(rotation=30,ha='right')
plt.bar(country.Country_name,country.Perceptions_of_corruption)
```

```python
 #corruption vs happiness
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x=df.Happiness_score,y=df.Perceptions_of_corruption,hue=df.Regional_indicator,s=200)
plt.legend(loc='lower left',fontsize='14')
plt.xlabel('Happiness score')
plt.ylabel('Corruption')
```

# Conclusion


1. Happiness Score and GDP:

There is a positive correlation between a country's Happiness Score and its GDP per capita. Countries with higher GDP tend to have higher happiness scores.
Regions like Western Europe and North America and ANZ have higher GDP per capita and, consequently, higher happiness scores.

2. Regional Differences:

Regional variations play a significant role in the Happiness Score. Different regions have different levels of happiness, with Western Europe and North America and ANZ generally having higher scores.
This suggests that regional factors, such as culture, social support, and government policies, can influence happiness.

3. Corruption Index:

The perception of corruption varies across regions. Some regions have a higher perception of corruption than others.
It's essential for countries to address corruption issues to improve their citizens' overall well-being and happiness.

4. Life Expectancy:

Life expectancy is positively correlated with happiness. Countries with higher happiness scores tend to have longer life expectancies.
Access to healthcare and overall well-being is crucial for happiness.

5. Freedom and Happiness:

There doesn't appear to be a strong linear correlation between freedom to make life choices and happiness score when considering all countries. However, regional variations exist.
The relationship between freedom and happiness may be influenced by other factors, such as social support and economic stability.

6. Top and Bottom Countries:

The top 10 happiest countries have high life expectancy and other contributing factors to happiness.
The bottom 10 countries may benefit from addressing factors like corruption and improving life expectancy to enhance overall happiness.

# Suggestions


1. Economic Development: To improve happiness, countries should focus on economic development and increasing GDP per capita. This can be achieved through various economic policies and investments in key sectors.

2. Regional Policies: Governments should consider regional differences and implement policies tailored to their specific cultural and social contexts.

3. Anti-Corruption Measures: To increase happiness, governments should work on reducing corruption and enhancing transparency in public administration.

4. Healthcare: Investing in healthcare infrastructure and ensuring access to quality healthcare can contribute to higher life expectancy and, consequently, happiness.

5. Social Support: Encouraging social support systems, such as strong family bonds and community networks, can enhance happiness.

6. Personal Freedom: While personal freedom is essential, it may not be the sole factor influencing happiness. A holistic approach to improving well-being should consider various aspects of citizens' lives.
